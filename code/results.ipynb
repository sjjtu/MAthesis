```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# acc, prec, rec, f1
baseline = [0.92, 0.9, 0.94,0.92]

aemerf = [0.83, 0.85, 0.79, 0.82]
aedpmerf_eps1 = [0.82, 0.81, 0.86, 0.83]
aedpmerf_eps05 = [0.81, 0.8, 0.82, 0.81]
aedpmerf_eps01 = [0.76, 0.75, 0.77, 0.76]
aedpmerf_eps001 = [0.59, 0.58, 0.59, 0.59]

rtsgan = [0.76, 0.76, 0.75, 0.75]
rtsgan_eps35 = [0.74, 0.74, 0.75, 0.74]
rtsgan_eps5 = [0.56, 0.56, 0.57, 0.56]
```

```python
plt.figure(figsize=(10,10))
titles = ["Accuracy", "Precision", "Recall", "F1"]
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.title(titles[i])
    plt.hlines(baseline[i], 0, 1, linestyle="--", color="green")
    plt.hlines(aemerf[i],0 , 1, linestyle="--", color="red")
    plt.ylim([0.5, 1.0])
    plt.plot([0.001, 0.01, 0.5, 1.0], [aedpmerf_eps001[i], aedpmerf_eps01[i], aedpmerf_eps05[i], aedpmerf_eps1[i]], marker="o")
    plt.legend(["baseline", "AE-MERF", "AE-DPMERF"], loc="lower right")
    plt.xlabel("Epsilon")
    plt.ylabel("Utility")
    plt.tight_layout()
```

```python
plt.figure(figsize=(10,10))
titles = ["Accuracy", "Precision", "Recall", "F1"]
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.title(titles[i])
    plt.hlines(baseline[i], 0, 35, linestyle="--", color="green")
    plt.hlines(rtsgan[i],0 , 35, linestyle="--", color="red")
    plt.ylim([0.5, 1.0])
    plt.plot([5, 35], [rtsgan_eps5[i], rtsgan_eps35[i]], marker="o")
    plt.legend(["baseline", "RTSGAN", "DP-RTSGAN"], loc="lower right")
    plt.xlabel("Epsilon")
    plt.ylabel("Utility")
    plt.tight_layout()
```
